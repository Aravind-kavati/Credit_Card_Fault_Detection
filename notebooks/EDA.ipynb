```python
 ## default payment next month dependant feature in the data
```

```python
X=data.drop(labels=['default payment next month'],axis=1)
```

```python
y=data['default payment next month']
```

```python
## these is a classification problem so we use classifier algorithms
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y,random_state=20,test_size=0.25)
```

```python
X_train.size,X_test.size
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random=RandomForestClassifier()
```

```python
## before model fitting we need to scal the data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_trained_scaled=scaler.fit_transform(X_train)
## to avoid the data leakage we do not use fit_transform((x-u)/sigma))
X_test_scaled=scaler.transform(X_test)
```

```python
random.fit(X_trained_scaled,Y_train)
```

```python
Y_pred=random.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,Y_pred))
```

```python
y
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gb=GaussianNB()
```

```python
gb.fit(X_trained_scaled,Y_train)
```

```python
Y_pred1=gb.predict(X_test)
```

```python
print(accuracy_score(Y_test,Y_pred1))
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_trained_scaled, Y_train)
Y_pred3=clf.predict(X_test_scaled)
print(accuracy_score(Y_test,Y_pred3))

```

```python
## random forest classifier
param_grid_rf = { "n_estimators": [50,100,130,200],
                    "max_depth": range(3,11,1),
                    "random_state":[0,50,100],
                     "criterion":['gini',"entropy"]
                     }
```

```python
from sklearn.model_selection import GridSearchCV
gs1=GridSearchCV(random,param_grid=param_grid_rf,cv=5,verbose=3)
```

```python
gs1.fit(X_trained_scaled,Y_train)
```

```python
gs1.best_params_
```

```python
from sklearn.ensemble import RandomForestClassifier
rs1=RandomForestClassifier(criterion='gini',n_estimators= 1200,random_state=0)
rs1.fit(X_trained_scaled,Y_train)
y_pred3=rs1.predict(X_test_scaled)
accuracy_score(Y_test,y_pred3)*100
```
